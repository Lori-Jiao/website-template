```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("STTotal_Victims_States_District_Court_Case__data.csv")
```

```python
df.head
```

```python
df.dtypes
```

```python
byage = df.groupby(["Age of Victims"])["Record ID"].count().sort_values(ascending=False).to_frame()
```

```python
byage.plot.pie (y="Record ID",title="Victim Age Distribution of Sex Trafficking Cases in California from 2003 to 2023 ")
```
